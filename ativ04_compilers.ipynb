```python
import re

def tokenize(texto):
  regexp = r'[a-zA-Z_][a-zA-Z0-9_]*|\d+|[=+\-*]'

  tokens = []
  tabela = {}
  contador = 1


  for match in re.finditer(regexp, texto):
    lexema = match.group()

    if re.match(r'[a-zA-Z_][a-zA-Z0-9_]*', lexema):
      if lexema not in tabela:
        tabela[lexema] = contador
        contador += 1
      tokens.append(f"(id,{tabela[lexema]})")

    elif re.match(r'\d+', lexema):
      tokens.append(f"({lexema})")
    else:
      tokens.append(f"({lexema})")


  return tokens, tabela

string = "position = initial + rate * 60"
tokens, tabela = tokenize(string)

print(" ".join(tokens))
```
